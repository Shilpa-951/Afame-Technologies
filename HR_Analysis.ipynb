```python
import pandas as pd
# Load the dataset
df = pd.read_csv("HR Data.csv")
# Remove unnecessary columns
columns_to_drop = ['EmployeeCount', 'StandardHours', 'Over18', 'EmployeeNumber']
df_cleaned = df.drop(columns=columns_to_drop)
```

1. Removing Unnecessary Columns

```python
columns_to_drop = ['EmployeeCount', 'StandardHours', 'Over18', 'EmployeeNumber']
df_cleaned = df.drop(columns=columns_to_drop)
```

```python
df.head()
```

2.Rename columns


```python
# Rename columns (example of a few columns, adjust as necessary)
rename_columns = {
    'Attrition': 'Employee Attrition',
    'BusinessTravel': 'Travel Frequency',
    'DailyRate': 'Daily Rate',
    'DistanceFromHome': 'Distance From Home',
    'EducationField': 'Field of Education',
    'JobRole': 'Job Role'
}
df_cleaned.rename(columns=rename_columns, inplace=True)
df_cleaned.head()
```

3. Eliminating Redundant Entries

```python
# Eliminate redundant entries
df_cleaned.drop_duplicates(inplace=True)
```

4. Sanitizing Specific Columns

```python
# Sanitize specific columns (ensuring categorical consistency)
df_cleaned['Employee Attrition'] = df_cleaned['Employee Attrition'].str.strip().str.capitalize()
df_cleaned['OverTime'] = df_cleaned['OverTime'].str.strip().str.capitalize()
```

5. Eliminating NaN Values

```python
# Check for null values in the dataset
null_values = df.isnull().sum()
null_values
```

```python
#Handle missing values by dropping rows with NaN values
df_cleaned.dropna(inplace=True)
df_cleaned
```


Here are a few additional changes that could be necessary:
Ensure Correct Data Types: Ensure columns have appropriate data types.
Normalize/Standardize Numerical Columns: Normalize numerical columns if preparing for machine learning.
Categorical Encoding: Convert categorical columns to numeric using one-hot encoding.

```python
# Ensure correct data types
df_cleaned['Age'] = df_cleaned['Age'].astype(int)
df_cleaned['Daily Rate'] = df_cleaned['Daily Rate'].astype(float)
df_cleaned['Distance From Home'] = df_cleaned['Distance From Home'].astype(int)
df_cleaned['Education'] = df_cleaned['Education'].astype(int)
df_cleaned['EnvironmentSatisfaction'] = df_cleaned['EnvironmentSatisfaction'].astype(int)
df_cleaned['HourlyRate'] = df_cleaned['HourlyRate'].astype(float)
df_cleaned['JobInvolvement'] = df_cleaned['JobInvolvement'].astype(int)
df_cleaned['JobLevel'] = df_cleaned['JobLevel'].astype(int)
df_cleaned['JobSatisfaction'] = df_cleaned['JobSatisfaction'].astype(int)
df_cleaned['MonthlyIncome'] = df_cleaned['MonthlyIncome'].astype(float)
df_cleaned['MonthlyRate'] = df_cleaned['MonthlyRate'].astype(float)
df_cleaned['NumCompaniesWorked'] = df_cleaned['NumCompaniesWorked'].astype(int)
df_cleaned['PercentSalaryHike'] = df_cleaned['PercentSalaryHike'].astype(int)
df_cleaned['PerformanceRating'] = df_cleaned['PerformanceRating'].astype(int)
df_cleaned['RelationshipSatisfaction'] = df_cleaned['RelationshipSatisfaction'].astype(int)
df_cleaned['StockOptionLevel'] = df_cleaned['StockOptionLevel'].astype(int)
df_cleaned['TotalWorkingYears'] = df_cleaned['TotalWorkingYears'].astype(int)
df_cleaned['TrainingTimesLastYear'] = df_cleaned['TrainingTimesLastYear'].astype(int)
df_cleaned['WorkLifeBalance'] = df_cleaned['WorkLifeBalance'].astype(int)
df_cleaned['YearsAtCompany'] = df_cleaned['YearsAtCompany'].astype(int)
df_cleaned['YearsInCurrentRole'] = df_cleaned['YearsInCurrentRole'].astype(int)
df_cleaned['YearsSinceLastPromotion'] = df_cleaned['YearsSinceLastPromotion'].astype(int)
df_cleaned['YearsWithCurrManager'] = df_cleaned['YearsWithCurrManager'].astype(int)

# Normalize/Standardize numerical columns (example for 'MonthlyIncome')
df_cleaned['MonthlyIncome'] = (df_cleaned['MonthlyIncome'] - df_cleaned['MonthlyIncome'].mean()) / df_cleaned['MonthlyIncome'].std()

# Convert categorical columns to numeric if preparing for machine learning
df_cleaned = pd.get_dummies(df_cleaned, columns=['Employee Attrition', 'Travel Frequency', 'Department', 'Field of Education', 'Gender', 'Job Role', 'MaritalStatus', 'OverTime'])

df_cleaned.head()
```
